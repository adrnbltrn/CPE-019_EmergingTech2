Question:  
**Notice at the left-to-right diagonal in the correlation table generated above. Why is the diagonal filled with 1s? Is that a coincidence? Explain.**

Answer:  
This is because the diagonal indicates the correlation of a variable with itself, which always has perfect value. The off-diagonal values have no interpretable meaning and are therefore usually eliminated from the correlation matrix.

Question:  
**Still looking at the correlation table above, notice that the values are mirrored; values below the 1
diagonal have a mirrored counterpart above the 1 diagonal. Is that a coincidence? Explain.**

Answer:  
This is because of the nature that correlation matrices are always completely symmetrical. Each cell in the matrix represents a correlation between two variables, and because there is no difference between A’s relationship with B compared to that of B on A, values are symmetrical across the diagonal.

Question:  
**Many variable pairs present correlation close to zero. What does that mean?**

Answer:  
When the majority of variable pairs show a correlation value approaching zero, it indicates that there is no linear relationship between those variables. A correlation coefficient close to 0 implies a weak or no relationship at all.


Question:  
**Why separate the genders?**

Answer:  
Splitting male and female data from a dataset is very necessary, as it would help in conducting gender specific analyses and comparisons. In investigating the relationship between variables, such as height and weight, it may be beneficial to investigate whether these relationships are different for males or females. Splitting the data allows researchers to obtain and compare correlations, means, standard deviations or other statistics for each gender separately.
